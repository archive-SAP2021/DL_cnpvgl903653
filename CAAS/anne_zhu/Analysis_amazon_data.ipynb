```python
import json
import pandas as pd
import time

def parse(path):
    with open(path) as file:
        for l in file:
            yield eval(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')
```

```python
path = "../resource/amazon_data/metadata.json"
start = time.time()
data = getDF(path)
end = time.time()
print(data.head(300))
```

```python
print('total loading time {}'.format(end- start))
```
